```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import cv2
import pandas as pd
```

```python
im = cv2.imread('./img.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
print im.shape

img = cv2.resize(im, (320, 240))
print img.shape
plt.figure(0)
plt.axis("off")
plt.imshow(img)
plt.show()
```

```python
image_vector = img.reshape((img.shape[0] * img.shape[1], img.shape[2]))
print image_vector.shape

labels=[]
labels.append('apple')
labels
```

```python
fruit_img =[]
fruit_img.append(image_vector)
```

```python
fruit_img=np.asarray(fruit_img)
print fruit_img.shape
np.save('./fruit_data',fruit_img)
```
